```python
import pandas as pd
import numpy as np

# Generate random product IDs
product_ids = np.arange(1, 1001)

# Generate random product names
product_names = ['Product ' + str(i) for i in product_ids]

# Generate random product prices
product_prices = np.random.randint(10, 100, size=len(product_ids))

# Generate random product categories
product_categories = ['Electronics', 'Clothing', 'Home Goods', 'Beauty', 'Toys']
categories_list = np.random.choice(product_categories, size=len(product_ids))

# Generate random customer IDs
customer_ids = np.arange(1, 501)

# Generate random customer names
customer_names = ['Customer ' + str(i) for i in customer_ids]

# Generate random customer email addresses
customer_emails = ['customer' + str(i) + '@email.com' for i in customer_ids]

# Generate random customer order dates
start_date = pd.to_datetime('2019-01-01')
order_dates = start_date + pd.to_timedelta(np.random.randint(0, (pd.to_datetime('2023-10-04') - start_date).days, size=len(customer_ids)), 'D')

# Generate random order amounts
order_amounts = np.random.randint(50, 200, size=len(customer_ids))

# Generate random order statuses
order_statuses = ['Pending', 'Processing', 'Shipped', 'Delivered']
status_list = np.random.choice(order_statuses, size=len(customer_ids))

# Create a DataFrame to store the generated data
data = {'product_id': product_ids,
        'product_name': product_names,
        'product_price': product_prices,
        'product_category': categories_list,
        'customer_id': np.repeat(customer_ids, 2),  # Repeat to match the length
        'customer_name': np.repeat(customer_names, 2),  # Repeat to match the length
        'customer_email': np.repeat(customer_emails, 2),  # Repeat to match the length
        'order_date': np.repeat(order_dates, 2),  # Repeat to match the length
        'order_amount': np.repeat(order_amounts, 2),  # Repeat to match the length
        'order_status': np.repeat(status_list, 2)}  # Repeat to match the length

df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('product_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Visualization 1: Histogram of Product Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['product_price'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Product Prices')
plt.xlabel('Product Price')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Bar Chart of Product Categories
plt.figure(figsize=(12, 6))
sns.countplot(x='product_category', data=df, palette='viridis')
plt.title('Bar Chart of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 3: Line Plot of Order Amounts Over Time
plt.figure(figsize=(12, 6))
df['order_date'] = pd.to_datetime(df['order_date'])
sns.lineplot(x='order_date', y='order_amount', data=df, estimator='sum', ci=None)
plt.title('Line Plot of Order Amounts Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Order Amount')
plt.show()

# Visualization 4: Pie Chart of Order Status Distribution
plt.figure(figsize=(8, 8))
df['order_status'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightgreen', 'lightblue'])
plt.title('Pie Chart of Order Status Distribution')
plt.ylabel('')
plt.show()

# Visualization 5: Scatter Plot of Product Prices vs. Order Amounts
plt.figure(figsize=(10, 6))
sns.scatterplot(x='product_price', y='order_amount', data=df, color='coral')
plt.title('Scatter Plot of Product Prices vs. Order Amounts')
plt.xlabel('Product Price')
plt.ylabel('Order Amount')
plt.show()

# Visualization 6: Boxplot of Order Amounts by Product Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='product_category', y='order_amount', data=df, palette='Set3')
plt.title('Boxplot of Order Amounts by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Order Amount')
plt.xticks(rotation=45)
plt.show()

# Visualization 7: Heatmap of Correlation Matrix
plt.figure(figsize=(10, 8))
correlation_matrix = df[['product_price', 'order_amount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()

# Visualization 8: Count Plot of Order Status by Product Category
plt.figure(figsize=(12, 6))
sns.countplot(x='product_category', hue='order_status', data=df, palette='husl')
plt.title('Count Plot of Order Status by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Order Status')
plt.show()
```
